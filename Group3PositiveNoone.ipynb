```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Define Filepath because macintosh hates me otherwise
file_path = '/Users/megnoone/Desktop/store_data.xlsx'
```

```python
#Read Excel File into matplotlib program
df = pd.read_excel(file_path, sheet_name='store_data')
```

```python
#define columns just in case we try something tricky
columns_to_use = ['Category', 'Profit',]
```

```python
#filter our information pool...eliminte to see true profits
df_filtered = df.loc[df['Profit'] > 0]
```

```python
#plot bar chart
plt.bar(df_filtered['Category'], df_filtered['Profit'], color='green')
plt.xlabel('Product Category')
plt.ylabel('Profit in Dollars')
plt.title('Profit in by Product Category')
plt.show()
```
