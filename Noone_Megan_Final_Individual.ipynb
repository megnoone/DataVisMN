```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Define filepath because Mac coding isn't great

data = pd.read_csv('/Users/megnoone/Desktop/School Shit/Grad School Stuff/Spring 2023/greenhouse-gas-emissions-industry-and-household-September-2022-quarter.csv')
```

```python
filtered_data = data[data['Period'].str.contains('Q1')]
```

```python
plt.bar(filtered_data['Period'], filtered_data['Data_value'])
plt.xlabel('Years')
plt.ylabel('Kilotonnes of CO2 Equivalents')
plt.title('Greenhouse gases emitted during Q1 of 2014-2022')
plt.show()
```
